## To get started with using a notebook in Beam
1. Run the command `notebook` in your Beam development session 
2. You can then connect to `http://0.0.0.0:2326/notebook/` in your browser or in VSCode
   1. Note: making changes in the browser notebook won't change your files locally

## Connect to Jupyter kernel in VSCode
1. Follow the steps above
2. Go to VSCode command palette and type in `Jupyter: Specify Jupyter Server for connections`
3. Click `Existing` and enter `http://0.0.0.0:2326/notebook/` for the input
4. At the top right of the Jupyter notebook, change your `kernel` to the (ipykernel) that is in a remote session

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```
